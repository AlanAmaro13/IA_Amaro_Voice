# Renombrador de archivos

Para almacenar los archivos en la base de datos utilizo el número que contiene el nombre del audio para emparejarlo con su texto asociado. Es fácil ver que esto no es del todo conveniente, dado que si tengo una carpeta con audios enumerados del 1 al 50 y muevo todos los archivos, ahora cuando grabe una serie de nuevos audios el contador empezará de nuevo desde 1. 


Para solucionar este incoveniente he diseñado este programa, su funcionamiento es el siguiente: Le das al audio el número a partir de cual quierer renombrar a los archivos, por decir 150, y buscara el audio con el numero 1 y le asignará el 151, de manera que ahora puedes mantener el indexado estándar sin necesidad de cambiarlos manualmente. 

```python
# Este programa solicita el numero nuevo de indexado y la ruta de archivos
from pathlib import Path
import os 
import re

def number_extraction(string): # Definimos una funcion para extraer numeros de un string
    numbers = re.findall(r'\d+', str(string)) # Usamos RegEx, asumimos solo un numero en cada str
    return int(numbers[0]) # Retornamos el numero entero


index = int(input('Ingrese el nuevo numero de index antiguo: '))
print("\n")

path = Path(input("Ingrese la ruta de los archivos: "))
print('\n')

list_archivos = os.listdir(path)

for nombre_archivo in list_archivos:
    ruta_archivo = (os.path.join(path, nombre_archivo)).replace('/', '\\')
    numero = number_extraction(nombre_archivo) 
    new_number = numero + index
    new_name = nombre_archivo[:-4].replace(str(numero), str(new_number) )
    new_name = new_name + ".m4a"
    new_path = (os.path.join(path, new_name)).replace('/', '\\')
    
    try:
        os.rename(ruta_archivo, new_path)
        print(f"Archivo: {numero} renombrado a: {new_number}\nEn {new_path}\n\n")
        
    except FileNotFoundError:
        print("Archivo no encontrado")
        
    
    
    
```
