# Nota: Mismos numeros para el texto y los audios

```python
import os # Para el manejo de rutas de archivos
import re # Para extraer los numeros de strings

import unicodedata

def quitar_acentos(texto):
    return ''.join(c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn')


def number_extraction(string): # Definimos una funcion para extraer numeros de un string
    numbers = re.findall(r'\d+', str(string)) # Usamos RegEx, asumimos solo un numero en cada str
    return int(numbers[0]) # Retornamos el numero entero

def extraer_palabras(string):
    # Utiliza una expresión regular para encontrar todas las palabras
    palabras = re.findall(r'\b\w+\b', string)
    return palabras[1]

def extraer_palabras_completas(string):
    # Utiliza una expresión regular para encontrar todas las palabras
    final_string = ""
    palabras = re.findall(r'\b\w+\b', string)
    for i in range(len(palabras)):
        if i == 0:
            pass
        else:
            final_string += palabras[i] + " "
            
    return final_string

txt = str(input("Ingresa el texto que contiene todas las palabras: \n \n"))  # entrada 
#txt = "1. Perro, 2. Gato, 3. Casa, 4. Árbol, 5. Sol, 6. Luna, 7. Estrella, 8. Coche, 9. Avión, 10. Barco, 11. Libro, 12. Escuela, 13. Computadora, 14. Teléfono"

# Manejo del texto:
txt_list = txt.split(", ") # Separamos cada mensaje por las comas
inicial_number = number_extraction(txt_list[0]) # Obtenemos el numero inicial de la lista
print("\n\n Texto almacenado y separado correctamente...\n Numero inicial: ", inicial_number , "\n\n")

print("\nIniciando limpieza del texto...\n")
new_txt_list = []
for word in txt_list: # Verificamos la entrada de la lista
    word = str(quitar_acentos(word)).lower() # Removemos acentos y hacemos minuscula
    new_txt_list.append(word)
    print("\n", word, "\n")


del txt_list
txt_list = new_txt_list
print("\n La cantidad de palabras ingresadas es: " , len(new_txt_list), "\n")
    
# Ingresamos la ruta de los audios enumerados 
path = str(input("\n\n Ingrese la ruta en la cual se contienen los audios enumerados: \n\n"))
path = path.replace('\\', '/')

list_audios = os.listdir(path) # Obtenemos la ruta de los audios
list_paths = [] # Obtenemos una lista de nombre, texto, nombre_audio, ruta

print(len(txt_list))

for nombre_archivo in list_audios: # Para cada archivo en la ruta dada
    ruta_archivo = (os.path.join(path, nombre_archivo)).replace('/', '\\')  # Cambia el formato de path
    numero = number_extraction(nombre_archivo) # Extrae el numero de cada archivo
    txt_in = extraer_palabras_completas( txt_list[numero - 1] )
    list_paths.append( [numero, txt_in, nombre_archivo, ruta_archivo] )
    
# Obtenemos una lista que incluye el texto del audio    
    
print("\n\n Los archivos se han enumerado de la siguiente forma:")    
for i in range(len(list_paths)):
    
    print(f'''
    Numero: {list_paths[i][0]}
    
    Texto: {list_paths[i][1]}
    
    Nombre: {list_paths[i][2]}
    
    Path: {list_paths[i][3]}
    
    
    ''')
    
    #print(f"\n Archivo: {nombre_archivo} | Ruta: {ruta_archivo} \n")
```

De esta forma obtenemos una lista que con el siguiente formato:

    list_path = [numero, txt_in, nombre_archivo, ruta_archivo]  

Donde el numero sera el id de la database, text_in es el texto que se menciona en cada audio, 
nombre del archivo es el nombre de la grabacion y ruta_achivo es claramente la ruta del audio.

```python
import os
from pydub import silence, AudioSegment
from pathlib import Path

AudioSegment.converter = r"C:\\Users\\blume\\ffmpeg-2024-03-25-git-ecdc94b97f-essentials_build\\bin\\ffmpeg.exe"
AudioSegment.ffprobe = r"C:\\Users\\blume\\ffmpeg-2024-03-25-git-ecdc94b97f-essentials_build\\bin\\ffprobe.exe"


def conversor_audio_bytes(lista): # Recibe list_path
    audio_path = Path(str(lista[3]))  # recibe path del audio
    audio = AudioSegment.from_file(audio_path, format = 'm4a')
    audio_binary = audio.export(format = 'mp3').read()
    
    audio_document = { # Genera el documento que se agregara a la db
        '_id' : lista[0],
        'Text': lista[1],
        'File_Name': lista[2],
        'Path' : lista[3],
        'data:' : audio_binary
    }
    
    return audio_document
```

```python
list_add = []
for item in list_paths:
    list_add.append(conversor_audio_bytes(item))

x = audios.insert_many(list_add)
```
